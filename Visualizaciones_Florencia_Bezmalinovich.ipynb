Para comenzar consult√© el tipo de datos que trae el dataset por defecto, luego defin√≠ las consultas que los 3 gr√°ficos deb√≠an resolver:

1¬∞ ¬øCual es la hora pico de vuelos en el territorio argentino?

```python
#Tipo de datos que trae el dataset por defecto
df_vuelo.dtypes
```

```python
#Gr√°fico 1: ¬øCual es la hora pico de vuelos en el territorio argentino? 
#1¬∞ Setear el campo hora que originalmente es un objeto como datetime: 
df_vuelo['Hora UTC'] = pd.to_datetime(df_vuelo['Hora UTC'])
```

```python
df_vuelo.dtypes
```

```python
#2¬∞ Setear la columna Hora UTC como indice.
df_vuelo.set_index('Hora UTC', inplace=True)
```

```python
df_vuelo.head()
```

```python
#3¬∞ Crear una variable que cuento los vuelos por horas.
flights_per_hour = df_vuelo.resample('H').count()
```

```python
import matplotlib.dates as mdates
```

```python
#4¬∞ Gr√°fico 1 con Matplotlib
fig, ax = plt.subplots(figsize=(18, 4))
plt.plot(flights_per_hour.index, flights_per_hour,flights_per_hour['Index'])
plt.xlabel('Horas')
plt.ylabel('Cantidad de vuelos')
plt.title('Count of Flights per Hour')

hours = mdates.HourLocator(interval=1)  
formatter = mdates.DateFormatter('%H:%M') 

ax.xaxis.set_major_locator(hours)
ax.xaxis.set_major_formatter(formatter)

plt.show()
```

```python
from dateutil import parser
```

```python
#4¬∞ Gr√°fico 1 con Seaborn
#sns.lineplot(data=flights_per_hour, x=flights_per_hour.index, y='Index')
plt.xlabel('Hour')
plt.ylabel('Count of Flights')
plt.title('Count of Flights per Hour')

# Create plot using dateplot() function
sns.set_style('whitegrid')
ax = sns.lineplot(data=flights_per_hour, x=flights_per_hour.index, y='Index')
ax.set_xlabel('Time')

# Format x-axis labels to show only the hour component of dates
labels = [item.get_text() for item in ax.get_xticklabels()]
labels = [parser.parse(label).strftime('%H:%M') for label in labels]
ax.set_xticklabels(labels)

# Show plot
sns.despine()
plt.show()
```

```python
#5¬∞ Quitar a la hora como Index.
df_vuelo.reset_index('Hora UTC', inplace=True)
```

```python
df_vuelo.head()
```

```python
#Gr√°fico 2: ¬øCual es el d√≠a de la semana donde hay m√°s aeronaves sobrevolando el territorio nacional?
#1¬∞ Formatear la fecha en datetime.
df_vuelo['Fecha UTC'] = pd.to_datetime(df_vuelo['Fecha UTC'])
```

```python
df_vuelo.head()
```

```python
#2¬∞ Crear variable D√≠a con el nombre del d√≠a de la semana correspondiente a esa fecha.
df_vuelo['D√≠a'] = df_vuelo['Fecha UTC'].dt.day_name()
```

```python
df_vuelo.head()
```

```python
#3¬∞ Crear un dataframe que agrupe los vuelos por los d√≠as de la semana
flights_per_day = df_vuelo.groupby('D√≠a')['Index'].count()
```

```python
flights_per_day.head()
```

```python
#4¬∞ Gr√°fico 2 con Matplotlib
plt.bar(flights_per_day.index, flights_per_day.values)
plt.xlabel('Day of the Week')
plt.ylabel('Count of Flights')
plt.title('Count of Flights per Day of the Week')
plt.show()
```

```python
flights_per_day = df_vuelo.groupby('D√≠a')['Index'].count().reset_index()
```

```python
#4¬∞ Gr√°fico 2 con Seaborn
sns.barplot(data=flights_per_day, x='D√≠a', y='Index')
plt.xlabel('Day of the Week')
plt.ylabel('Count of Flights')
plt.title('Count of Flights per Day of the Week')
plt.show()
```

```python
#5¬∞ Validaci√≥n de gr√°fico 2.
import statistics
mode_day = statistics.mode(df_vuelo['D√≠a'])
print("La moda de los datos es:", mode_day)
```

```python
# Grafico 3: ¬øCuales son las 10 aerolineas con m√°s vuelos dom√©sticos en el pa√≠s (Matplotlib) y cuales con m√°s vuelos internacionales (Seaborn)?
#1¬∞ Crear un dataframe que filtre los vuelos por su naturaleza de dom√©stico o internacional.
df_vuelos_domesticos = df_vuelo.loc[df_vuelo['Clasificaci√≥n Vuelo'] =='Dom√©stico']
df_vuelos_internacionales = df_vuelo.loc[df_vuelo['Clasificaci√≥n Vuelo'] =='Internacional']
```

```python
#2¬∞ Crear el dataframe de vuelos dom√©sticos y otro por vuelos internacionales filtrandolo por aerolinea.
domestics_flights_per_airline = df_vuelos_domesticos.groupby('Aerolinea Nombre')['Index'].count()
#international_flights_per_airline = df_vuelos_internacionales.groupby('Aerolinea Nombre')['Index'].count()
```

```python
#3¬∞ Ordenar en cada dataframe las aerolineas por orden descendentes para que aparezcan las aerolineas con m√°s vuelos al inicio.
domestics_flights_per_airline = domestics_flights_per_airline.sort_values(ascending=False)
```

```python
#4¬∞ Filtrar las 10 primeras para mejor visualizaci√≥n.
domestics_flights_per_airline = (domestics_flights_per_airline[:5])
```

```python
#5¬∞ Gr√°fico 3 con Matplotlib
plt.barh(domestics_flights_per_airline.index, domestics_flights_per_airline.values)
plt.xlabel('Count of Flights')
plt.ylabel('Airline')
plt.title('Number of Domestics Flights per Airline')
plt.show()
```

```python
international_flights_per_airline = df_vuelos_internacionales.groupby('Aerolinea Nombre')['Index'].count().reset_index()
```

```python
international_flights_per_airline = international_flights_per_airline.sort_values(by='Index', ascending=False)
```

```python
international_flights_per_airline = (international_flights_per_airline[:5])
```

```python
#4¬∞ Gr√°fico 3 con Seaborn
sns.barplot(data=international_flights_per_airline, x='Index', y='Aerolinea Nombre', color='blue')
plt.xlabel('Count of Flights')
plt.ylabel('Airline')
plt.title('Number of International Flights per Airline')

plt.show()
```
